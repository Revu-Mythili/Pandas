```python
#correlation between columns
df.corr(numeric_only=True)
```

```python
#heatmap using seaborn 
#1 - highly correlated 0 - no correlation minus - negative correlation

sns.heatmap(df.corr(numeric_only = True), annot=True)
plt.rcParams['figure.figsize'] = (20,20)
plt.show()
```

```python
#grouping data
# use case - are there certain continents which are growing faster than others and in which ways
df.groupby('Continent').mean(numeric_only=True)
```

```python
df[df['Continent'].str.contains('Oceania')]
```

```python
df2 = df.groupby('Continent').mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
df2
```
